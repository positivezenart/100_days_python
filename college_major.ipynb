```python
import pandas as pd
```

```python
df =pd.read_csv('/content/salaries_by_college_major.csv')
```

```python
df.head()
```

```python
print(f"Total number of rows,{df.axes[0].size}")
print(f"Total number of columns,{df.axes[1].size}")
print(f"List of columns, {list(df.columns)}")
```

```python
def null_checker(df) :
  [print(f"{i}Column contains Null value") if df[i].isna().any() else print(f"{i} column has No null values") for i in df.columns ]
```

```python
df_null_free = df.dropna()
```

```python
df_null_free.tail()
```

```python
null_checker(df_null_free)
```

```python
df_null_free['Starting Median Salary'].idxmax()
df_starting_max=df_null_free.iloc[43]
print(f"{df_starting_max[0]} college major has the hight starting Median salary as {df_starting_max[1]}")
```

```python
df_null_free['Mid-Career Median Salary'].idxmax()
df_mid_max=df_null_free.iloc[8]
print(f"{df_mid_max[0]} college major has the hight Mid Career Median salary as {df_mid_max[1]}")
```

```python
df_null_free['Starting Median Salary'].idxmin()
df_starting_min=df_null_free.iloc[49]
print(f"{df_starting_min[0]} college major has the lowest starting Median salary as {df_starting_min[1]}")
```

```python
df_mid_min=df_null_free.iloc[df_null_free['Mid-Career Median Salary'].idxmin()]
print(f"{df_mid_min[0]} college major has the lowest Mid Career Median salary as {df_mid_min[1]}")
print(f"{df_mid_min[0]} college major can expect to earn upto {df_mid_min[4]}")
```

```python
diff_salary =df_null_free['Mid-Career 90th Percentile Salary'].subtract(df_null_free['Mid-Career 10th Percentile Salary'])
```

```python
df_null_free.insert(1,'diff_salary',diff_salary)
```

```python
df_null_free.head()
```

```python
low_risk = df_null_free.sort_values('diff_salary', ascending=False)
```

```python
low_risk[['Undergraduate Major','diff_salary']].head()
```

```python
low_risk[['Undergraduate Major','Mid-Career 90th Percentile Salary']].sort_values('Mid-Career 90th Percentile Salary',ascending=False).head()
```

```python
change_in_percent = (low_risk[['Mid-Career 10th Percentile Salary', 'Mid-Career 90th Percentile Salary']].pct_change(axis=1)['Mid-Career 90th Percentile Salary'] * 100).round(2)
df_null_free.insert(2,'perc_change',change_in_percent)
```

```python
df_null_free.head()
```

```python
change_in_perce = df_null_free.sort_values('perc_change',ascending=False)
```

```python
change_in_perce[['Undergraduate Major','Starting Median Salary','Mid-Career 90th Percentile Salary','perc_change']].head()
```
